```python
# import our modules that we are using
import matplotlib.pyplot as plt 
import numpy as np 

# create the vectors X and Y
x = np.array(range(100))
y = x ** 2

# create the plot
plt.plot(x,y,label='y = x**2')

# add a title
plt.title('My First Plot')

# add X and Y label
plt.xlabel('x axis')
plt.ylabel('y axis')

# add a grid
plt.grid(alpha=.4,linestyle='--')

# add a legend
plt.legend()

# show the plot
plt.show()
```
